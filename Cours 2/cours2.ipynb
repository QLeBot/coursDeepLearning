```python
# crer un modle squentiel avec 4 couches de neurones avec une activation tanh et une couche de sortie avec une activation sigmode
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

model2 = Sequential()
model2.add(Dense(4, input_shape=(4, ), activation='tanh'))
model2.add(Dense(1, activation='sigmoid'))
model2.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

EXERCICE 3

```python
model3 = Sequential()
model3.add(Dense(128, input_shape=(2, ), activation='tanh'))
model3.add(Dense(64, activation='tanh'))
model3.add(Dense(32, activation='tanh'))
model3.add(Dense(4, activation='softmax'))

model3.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
